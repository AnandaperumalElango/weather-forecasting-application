```python
import io
import tkinter as tk
import requests
from PIL import Image, ImageTk

# Create the GUI window
root = tk.Tk()
root.title("Weather Forecast")
#root.config(bg="blue")

# Create the GUI widgets
label_city = tk.Label(root, text="Enter city name:")
entry_city = tk.Entry(root)
button_search = tk.Button(root, text="Search",bg="sky blue")
label_space =tk.Label(root , text="  ")#change bg color
label_space1 =tk.Label(root , text="  ")#change bg color

label_temp = tk.Label(root, bg="black" ,fg="white")
label_desc = tk.Label(root ,bg="black" ,fg="white")
label_icon = tk.Label(root)
label_min =tk.Label(root,  bg="black" ,fg="white")
label_max =tk.Label(root, bg="black" ,fg="white")
label_humidity = tk.Label(root, bg="black" ,fg="white")

# Add the widgets to the window
label_city.pack()
entry_city.pack()
label_space.pack()
button_search.pack()
label_space1.pack()
label_temp.pack()
label_min.pack()
label_max.pack()
label_humidity.pack()

label_desc.pack()
label_icon.pack()

# Define the function to get the weather data
def get_weather():
    # Get the city name from the entry widget
    city = entry_city.get()

    # Make the API call to OpenWeatherMap
    #api_key = "a08371d145ef92d30a967b23e1b23454"
    api_key ="b9bc3662c4fb1351a6441fe8ae39b695"
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)
    weather_data = response.json()
    print(response)
    print(weather_data)
    # Parse the weather data
    temp = weather_data['main']['temp']
    mini_temp = weather_data['main']['temp_min']
    maxi_temp = weather_data['main']['temp_max']
    humidity =weather_data['main']['humidity']
    print(temp)
    desc = weather_data["weather"][0]["description"]
    icon_code = weather_data["weather"][0]["icon"]

    # Update the GUI labels with the weather data
    label_temp.config(text=f"Temperature: {temp}°C")
    label_desc.config(text=f"Description: {desc}")
    label_min.config(text=f"minimum temperature: {mini_temp}°C")
    label_max.config(text=f"maximum temperature: {maxi_temp}°C")
    label_humidity.config(text=f"Humidity:{humidity}%")
    # Load the weather icon
    icon_url = f"http://openweathermap.org/img/wn/{icon_code}.png"
    icon_response = requests.get(icon_url)
    icon_data = icon_response.content
    icon_image = ImageTk.PhotoImage(Image.open(io.BytesIO(icon_data)))
    label_icon.config(image=icon_image)
    label_icon.image = icon_image

# Bind the search button to the get_weather function
button_search.config(command=get_weather)

# Start the GUI event loop
root.mainloop()
```
